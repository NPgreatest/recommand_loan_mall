```python
def create_json_file(text_list, file_name="edu_test.json"):
    with open(file_name, "w") as file:
        for text in text_list:
            line_dict = {"edu": text}
            line_str = repr(line_dict) + "\n"
            file.write(line_str)

create_json_file(results)
"JSON文件已成功创建。"
```

```python
import mysql.connector
def connect_to_database():
    connection = mysql.connector.connect(
        host='127.0.0.1',
        user='root',
        password='root',
        database='mall'
    )
    return connection
connection = connect_to_database()
cursor = connection.cursor()

sql_str="SELECT ASIN FROM mall_goods_info"
cursor.execute(sql_str)
results = cursor.fetchall()
results=[res[0] for res in results]
results_set=set(results)
connection.close()
results_set
```

```python
import json
with open("Grocery_and_Gourmet_Food.json", "r") as source_file, open("shrink_review", "w") as target_file:
    for line in source_file:
        review_data = json.loads(line)
        if review_data.get("asin") in results_set:
            json.dump(review_data, target_file)
            target_file.write("\n")
"JSON文件已成功创建。"
```
